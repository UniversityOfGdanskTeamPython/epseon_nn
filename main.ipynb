```python
"""Heh."""
from __future__ import annotations

import matplotlib.pyplot as plt
import numpy as np
import numpy.typing as npt
```

${\displaystyle V(r)=D_{e}(1-e^{-\alpha (r-r_{0})})^{2}+V(r_{0})}$


```python
D_e = 5500.0  # Dissociation energy
r_e = 4.5  # Equilibrium bond distance
a = 0.5  # Controls the 'width' of the potential well
```

```python
# Range of internuclear distances
r = np.linspace(-2.0, 20, 16500, dtype=np.float32)

V = (D_e * (1 - np.exp(-a * (r - r_e), dtype=np.float32)) ** 2) - D_e

plt.plot(r, V)
plt.xlim(-0.0, 20.0)
plt.ylim(-7000.0, 1000.0)
plt.hlines([0], -10.0, 20.0, linestyles="--", color="red")
plt.grid(True)
```

```python
def get_1() -> tuple[npt.NDArray[np.float32], npt.NDArray[np.float32]]:
    """Generate Morse Potential vector."""
    r = np.linspace(-2.0, 20, 16500, dtype=np.float64)
    V = (D_e * (1 - np.exp(-a * (r - r_e), dtype=np.float64)) ** 2) - D_e
    return (r, V)
```

```python
vals = []

for _ in range(16384):
    vals.append(get_1())
```
